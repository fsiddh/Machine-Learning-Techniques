```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
X = df.drop('Survived', axis='columns')
y = df[['Survived']]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le_Sex = LabelEncoder()
X.Sex = le_Sex.fit_transform(X.Sex)
```

```python
X.head()
```

```python
X.Age
```

```python
count = 0
for i in range(len(X)):
    if X.Age.isnull()[i] == True:
        count = count + 1

print(count)
```

```python
import numpy as np

m = X.Age.mean(skipna = True)
X.Age = X.Age.replace(np.NaN, m)
```

```python
count = 0
for i in range(len(X)):
    if X.Age.isnull()[i] == True:
        count = count + 1

print(count)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
survival_tree = DecisionTreeClassifier(criterion='entropy', max_depth=4)
```

```python
survival_tree.fit(X_train, y_train)
```

```python
y_hat = survival_tree.predict(X_test)
```

```python
from sklearn import metrics

accuracy = metrics.accuracy_score(y_test, y_hat)
print(accuracy)
```
